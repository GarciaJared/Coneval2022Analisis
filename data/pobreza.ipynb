INDICADORES DE CARENCIAS SOCIALES (REZAGO EDUCATIVO)

```python
population = pd.read_csv(r'files/poblacion.csv')
pob= population.copy()
pob
```

```python
pob[['parentesco', 'edad', 'asis_esc','nivelaprob', 'gradoaprob', 'antec_esc','hablaind' ]]=pob[['parentesco', 'edad', 'asis_esc','nivelaprob', 'gradoaprob', 'antec_esc','hablaind']].apply(pd.to_numeric, errors='coerce')
```

```python
pob=pob[~((pob['parentesco'] >= 400) & (pob['parentesco'] <= 500) | 
          (pob['parentesco']>= 700) & (pob['parentesco'] <= 800))]
pob
```

```python
pob['anac_e'] = 2022 - pob['edad'].fillna(0)
pob
```

```python
pob['inas_esc'] =np.NAN
pob.loc[(pob['asis_esc'] ==1 , 'inas_esc')] = 0
pob.loc[(pob['asis_esc']==2, 'inas_esc')] = 1
pob
```

```python
pob['niv_ed'] = np.NAN
pob.loc[(pob['nivelaprob']<2) | ((pob['nivelaprob']==2) & (pob['gradoaprob']<6)), 'niv_ed'] = 0
pob.loc[((pob['nivelaprob'] == 2) & (pob['gradoaprob'] == 6)) |
           ((pob['nivelaprob'] == 3) & (pob['gradoaprob'] < 3)) |
           (((pob['nivelaprob'] == 5) | (pob['nivelaprob'] == 6)) & (pob['gradoaprob'] < 3) & (pob['antec_esc'] == 1)), 'niv_ed'] = 1
# Secundaria completa o media superior incompleta
pob.loc[((pob['nivelaprob'] == 3) & (pob['gradoaprob'] == 3)) |
           ((pob['nivelaprob'] == 4) & (pob['gradoaprob'] < 3)) |
           ((pob['nivelaprob'] == 5) & (pob['antec_esc'] == 1) & (pob['gradoaprob'] >= 3)) |
           ((pob['nivelaprob'] == 6) & (pob['antec_esc'] == 1) & (pob['gradoaprob'] >= 3)) |
           ((pob['nivelaprob'] == 5) & (pob['antec_esc'] == 2) & (pob['gradoaprob'] < 3)) |
           ((pob['nivelaprob'] == 6) & (pob['antec_esc'] == 2) & (pob['gradoaprob'] < 3)), 'niv_ed'] = 2
# Media superior completa o mayor nivel educativo
pob.loc[((pob['nivelaprob'] == 4) & (pob['gradoaprob'] == 3)) |
           ((pob['nivelaprob'] == 5) & (pob['antec_esc'] == 2) & (pob['gradoaprob'] >= 3)) |
           ((pob['nivelaprob'] == 6) & (pob['antec_esc'] == 2) & (pob['gradoaprob'] >= 3)) |
           ((pob['nivelaprob'] == 5) & (pob['antec_esc'] > 2)) |
           ((pob['nivelaprob'] == 6) & (pob['antec_esc'] > 2)) |
           ((pob['nivelaprob'] >= 7) & (~pob['nivelaprob'].isna())), 'niv_ed'] = 3
```

```python
print(pob['niv_ed'].describe())
```

```python
# 1. Tiene de tres a 21 años, no cuenta con la educación obligatoria y no asiste a un centro de educación formal.
criterio1 = (pob['edad'] >= 3) & (pob['edad'] <= 21) & (pob['niv_ed'] < 2) & (pob['inas_esc'] == 1)

# 2. Tiene 22 años o más, nació a partir del año 1998 y no ha terminado la educación obligatoria (media superior).
criterio2 = (pob['edad'] >= 22) & (pob['anac_e'] >= 1998) & (pob['niv_ed'] < 3)

# 3. Tiene 16 años o más, nació entre 1982 y 1997 y no cuenta con el nivel de educación obligatorio vigente en el momento en que debía haberlo cursado (secundaria completa).
criterio3 = (pob['edad'] >= 16) & (pob['anac_e'] >= 1982) & (pob['anac_e'] <= 1997) & (pob['niv_ed'] < 2)

# 4. Tiene 16 años o más, nació antes de 1982 y no cuenta con el nivel de educación obligatorio vigente en el momento en que debía haberlo cursado (primaria completa).
criterio4 = (pob['edad'] >= 16) & (pob['anac_e'] < 1982) & (pob['niv_ed'] < 1)
```

```python
total_pob = len(pob)
total_pobtotal_poblacion = len(pob)
porcentaje_criterio1 = criterio1.sum() / total_pob * 100
porcentaje_criterio2 = criterio2.sum() / total_pob * 100
porcentaje_criterio3 = criterio3.sum() / total_pob * 100
porcentaje_criterio4 = criterio4.sum() / total_pob * 100
total_pobtotal_poblacion = porcentaje_criterio1 + porcentaje_criterio2 + porcentaje_criterio3 + porcentaje_criterio4
```

```python
# Mostrar resultados
print(f"Porcentaje que cumple con el criterio 1: {porcentaje_criterio1:.2f}%")
print(f"Porcentaje que cumple con el criterio 2: {porcentaje_criterio2:.2f}%")
print(f"Porcentaje que cumple con el criterio 3: {porcentaje_criterio3:.2f}%")
print(f"Porcentaje que cumple con el criterio 4: {porcentaje_criterio4:.2f}%")
print(f"Total de la población: {total_pob}")
print(f"Total de la población que cumple con los criterios: {total_pobtotal_poblacion:.2f}%")
```

```python
results={
    'Criterio': ['Criterio 1', 'Criterio 2', 'Criterio 3', 'Criterio 4', 'total_pob', 'total_pobtotal_poblacion'],
    'Porcentaje': [porcentaje_criterio1, porcentaje_criterio2, porcentaje_criterio3, porcentaje_criterio4, total_pob, total_pobtotal_poblacion]
}
```

```python
# Exportar a un archivo JSON
with open('resultados.json', 'w') as json_file:
    json.dump(results, json_file)
```
